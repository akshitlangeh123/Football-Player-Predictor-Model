##Splitting the data into training and testing data (for classification of player positions)

```python
from sklearn.model_selection import train_test_split
X1_train,X1_test,y1_train,y1_test = train_test_split(X1,y1,test_size=0.25,random_state=0)
```

##Scaling the data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X1_train = sc.fit_transform(X1_train)
X1_test = sc.fit_transform(X1_test)
```

##Applying Classification Algorithms for finding the players position along with their accuracies

```python
from sklearn.neighbors import KNeighborsClassifier
classifier1 = KNeighborsClassifier(n_neighbors=55,metric='minkowski',p=2)
classifier1.fit(X1_train,y1_train)
```

```python
y1_pred1 = classifier1.predict(X1_test)
print(np.concatenate((y1_pred1.reshape(len(y1_pred1),1),y1_test.reshape(len(y1_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm1 = confusion_matrix(y1_test,y1_pred1)
print(cm1)
accuracy_score1 = accuracy_score(y1_test,y1_pred1)
print("{:2f}".format(accuracy_score1))
```

```python
from sklearn.svm import SVC
classifier2 = SVC(kernel='rbf',random_state=0)
classifier2.fit(X1_train,y1_train)
```

```python
y1_pred2 = classifier2.predict(X1_test)
print(np.concatenate((y1_pred2.reshape(len(y1_pred2),1),y1_test.reshape(len(y1_test),1)),1))
```

```python
cm2 = confusion_matrix(y1_test,y1_pred2)
print(cm2)
accuracy_score2 = accuracy_score(y1_test,y1_pred2)
print("{:2f}".format(accuracy_score2))
```

<i>rbf>linear>sigmoid (work) (accuracy)</i>

```python
from sklearn.naive_bayes import GaussianNB
classifier3 = GaussianNB()
classifier3.fit(X1_train,y1_train)
```

```python
y1_pred3 = classifier3.predict(X1_test)
print(np.concatenate((y1_pred3.reshape(len(y1_pred3),1),y1_test.reshape(len(y1_test),1)),1))
```

```python
cm3 = confusion_matrix(y1_test,y1_pred3)
print(cm3)
accuracy_score3 = accuracy_score(y1_test,y1_pred3)
print("{:2f}".format(accuracy_score3))
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier4 = DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier4.fit(X1_train,y1_train)
```

```python
y1_pred4 = classifier4.predict(X1_test)
print(np.concatenate((y1_pred4.reshape(len(y1_pred4),1),y1_test.reshape(len(y1_test),1)),1))
```

```python
cm4 = confusion_matrix(y1_test,y1_pred4)
print(cm4)
accuracy_score4 = accuracy_score(y1_test,y1_pred4)
print("{:2f}".format(accuracy_score4))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier5 = RandomForestClassifier(n_estimators=50,criterion='entropy',random_state=0)
classifier5.fit(X1_train,y1_train)
```

```python
y1_pred5 = classifier5.predict(X1_test)
print(np.concatenate((y1_pred5.reshape(len(y1_pred5),1),y1_test.reshape(len(y1_test),1)),1))
```

```python
cm5 = confusion_matrix(y1_test,y1_pred5)
print(cm5)
accuracy_score5 = accuracy_score(y1_test,y1_pred5)
print("{:2f}".format(accuracy_score5))
```
